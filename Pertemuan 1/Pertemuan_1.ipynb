Membuat Array

```python
import array as arr
# Array bilangan bulat
a = arr.array('i', [1, 2, 3])
# Array bilangan desimal
b = arr.array('d', [2.5, 3.2, 3.3])
print("Array a:", a)
print("Array b:", b)
```

Menambah Elemen

```python
import array as arr

a = arr.array('i', [1, 2, 3])
print("Sebelum insert:", a)
a.insert(1, 4)
print("Setelah insert:", a)
b = arr.array('d', [2.5, 3.2, 3.3])
print("Sebelum append:", b)
b.append(4.4)
print("Setelah append:", b)
```

Mengakses Elemen

```python
import array as arr
a = arr.array('i', [1, 2, 3, 4, 5, 6])
print("Access element is: ", a[0])
print("Access element is: ", a[3])
b = arr.array('d', [2.5, 3.2, 3.3])
print("Access element is: ", b[1])
print("Access element is: ", b[2])
```

Menghapus Elemen

```python
import array as arr

a = arr.array('i', [1, 2, 3, 2, 4])
print("Array awal:", a)

a.remove(2)
print("Setelah remove(2):", a)

a.pop()
print("Setelah pop():", a)

a.pop(1)
print("Setelah pop(1):", a)
```

Pemotongan Elemen Array

```python
import array as arr

a = arr.array('i', [1, 2, 3, 4, 5, 6, 7, 8])

print(a[3:8])     # indeks 3 sampai 7
print(a[5:])      # indeks 5 sampai akhir
print(a[:])       # seluruh array
print(a[ ::- 1])  # membalik array
```

Mengubah Elemen Array

```python
import array as arr

a = arr.array('i', [1, 2, 3, 4])

a[1] =10
a[3] =20

print("Setelah diubah:", a)
```

Operasi Pada Array

```python
# Menghitung Elemen (count)
import array as arr
a = arr.array('i', [1, 2, 2, 3, 2])
print("Jumlah angka 2:", a.count(2))
```

```python
# Membalik Array (reverse)
a = arr.array('i', [1, 2, 3, 4])
a.reverse()

print("Setelah reverse:", a)
```

```python
# Extend Array
# Menambahkan banyak elemen sekaligus.
import array as arr

a = arr.array('i', [1, 2, 3])
b = arr.array('i', [4, 5, 6])

print("Sebelum extend:", a)

a.extend(b)

print("Setelah extend:", a)
```
